```python
# filter ()
'''
filter(function, iterable)
'''
def even(num):
    return num % 2 == 0
```

```python
even(5)
```

```python
even(4)
```

```python
from random import random
lst = [int(random() * 20) for _ in range(10)]
lst
```

```python
list(filter(even,lst))
```

```python
list(filter(lambda n: n % 2 == 0,lst))
```

```python
list(map(lambda n: n % 2 == 0,lst))
```

```python
# using filter with a dictionary
# filter items where the value is greater than 5

d = {'apple': 3, 'banana': 7, 'cherry': 2, 'date': 9}

ft = dict(filter(lambda item: item[1] > 5, d.items()))
```

```python
ft
```

```python
d.items()
```

```python
d.keys()
```

```python
d.values()
```

```python
# Returns true if value of item is even

ft = dict(filter(lambda value: value[1] % 2 == 0, d.items()))
ft
```

```python
# filter items where the key starts with 'b'

ft = dict(filter(lambda item: item[0][0] == 'b', d.items()))
ft
```

```python
# Reduce Function reduce()
#reduce(fucntion, iterable[, initializer])

lst = list(range(5,51,5))
lst
```

```python
from functools import reduce
prod = reduce(lambda x,y: x * y, lst)
prod
```

```python
prod = reduce(lambda x,y: x * y, lst, 2)
prod
```

```python
minvalue = reduce(lambda x,y: x if x < y else y, lst)
minvalue
```

```python
maxvalue = reduce(lambda x,y: x if x > y else y, lst)
maxvalue
```

```python
s = reduce(lambda x,y: x + y, lst)
s
```
