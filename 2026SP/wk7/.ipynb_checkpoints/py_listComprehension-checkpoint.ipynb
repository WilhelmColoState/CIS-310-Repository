### Python List Comprehension

- lsit comphrehention mae it easy to create lists while performing sophisticated filtering, mapping, and conditional logic
on their members

-sipmle powerful tools that you can use in a wide cariety of situations

<code new_list = [expression ** for member in iterable**] </code>

<code> new_list = [expression (if conditional) **for member in iterable**]

```python
# create a list from a string
lst = []

for letter in 'hello':
    lst.append(letter)
    
lst
```

```python
# [expression for item in iterable]

lst = [letter for letter in 'hello']
```

```python
lst
```

```python
squares = [n**2 for n in range(1,10)]
squares
```

```python
#convert celsius to fahrenheit

c_lst = [4,10,20,25.5,15.8,0,14.9]

f_lst = [(9/5) * t + 32 for t in c_lst]

f_lst
```

```python
lst = [n for n in range(1, 16) if n % 2 ==0]

lst
```

```python
odds = [n for n in range(1, 16) if n % 2 !=0]
```

```python
odds
```

```python
s = "CIS310 communicate students with analytical thinking and practical skills"

# find vowels fromm a sentence

vowels = [l for l in s if l.lower() in 'aeiou']
vowels
```

```python
# new_list = [expression (if conditional) for member in iterable]

prices = [1.25, -9.45, 10.22, 3.78, -5.92, 1.16]

new_prices = [n if n > 0 else 0 for n in prices]
new_prices
```

```python
lst = [n for n in range(16) if n % 2 == 0]
lst
```

```python
words = ['Even' if v % 2 == 0 else 'Odd' for v in range(16)]
words
```

```python
# nested list comprehensions
squares = [num**2 for num in [ n for n in range(1,11) if n%2 == 0]]
squares
```

```python
# using set and dictionary comprehension
t = (n for n in range(11))
t
```

```python
s = {n**2 for n in range(1,11)}
s
```

```python
s = {n for n in range(10,21) if n % 2 == 0}
s
```

```python
# Dictionary comprejensions are similar,
# with additional reqiremnets of defining a key:

d = {n:n**2 for n in range(5,16)}
d
```

```python
d = {n:n**2 for n in range(5,16) if n % 2 !=0}
d
```

```python
# use of list comprehension in dictionary comprehension
cities = ['Austin', 'Tacoma', 'Topeka', 'Sacremento', 'Charlotte']

temps = {city: [t for t in range(40, 60, 5)] for city in cities}
temps
```

```python
from random import randint



temps = {city:randint(40, 100) for city in cities}
temps
```
