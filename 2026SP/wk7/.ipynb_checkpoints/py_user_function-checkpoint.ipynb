**Python Function Format**

    docstring (Document String) - This is where you descripe a function

```python
def welcome():
    print('Welcome to CIS310!')
```

```python
welcome()
```

```python
# function with parameter(s)
def greet(name):
    '''
    This function greets a person's name passed as a parameter.
    '''
    print(f'Hello {name}, welcome to CIS310!')
```

```python
greet('Bill')
```

```python
print(greet.__doc__)
```

```python
#using return
def avg(n1,n2,n3):
    return(n1 + n2 + n3) / 3
```

```python
avg(10, 5, 12)
```

```python
x,y,z = 5,10,8
print(f" The average of values: {x}, {y}, {z} is {round(avg(x,y,z),2)}")
```

```python
def full_name(fname, lname):
    return lname + ", " +fname
```

```python
print(full_name('Bill', 'Lange'))
```

```python
def min_two_odd(n1,n2):
    if n1 % 2 != 0 and n2 % 2 !=0:
        return min(n1, n2)
    else:
        return max(n1, n2)
```

```python
min_two_odd(93, 105)
```

```python
def is_prime(num):
    for n in range(2, num):
        if num % n == 0:
            return str(num) + ' is not prime'
        
    return str(num) + ' is prime'
```

```python
is_prime(9)
```
