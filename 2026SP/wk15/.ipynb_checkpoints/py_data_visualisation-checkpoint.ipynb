```python
# using format string, such as colors, markers and line styles
plt.plot(randn(30), 'ko--')
```

```python
plt.plot(randn(30) * 4, color = 'k', linestyle='dotted', marker='o')
```

```python
plt.plot(randn(30), drawstyle='steps-post', label='steps-post')
plt.plot(randn(30).cumsum(), 'k--', label='dashed-line')

plt.legend(loc='best') # loc
```

```python
#plot multiple columns
df = DataFrame(randn(500,2), columns = list('AB'))
df = df.cumsum()
```

```python
plt.plot(df['A'], 'k.', label="Dataset A")
plt.plot(df['B'], color='blue', label="Dataset B")

plt.legend(loc='best')
```

```python
#Ticks, Labels and legends
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(randn(1000).cumsum())
```

```python
ax.set_xticks(list(range(0,1001,250)))

ax.set_xticklabels(['one','two', 'three', 'four', 'five'], rotation = 30, fontsize=12)

ax.set_title("My FIrst Matplotlip Plot with Ticks")

ax.set_xlabel('Stages')
ax.set_ylabel('Values')

fig
```

```python
props = {'title': 'My First Matplotlib plot with Ticks',
        'xlabel': 'Stages',
        'ylabel': 'Values'}

ax.set(**props)
fig
```
