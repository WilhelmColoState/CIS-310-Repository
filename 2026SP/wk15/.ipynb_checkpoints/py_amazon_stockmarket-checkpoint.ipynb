```python
#count the nuber of numerical and non-numeric variables (columns)
from pandas.api.types import is_numeric_dtype

n = 0
for c in data.columns:
    if is_numeric_dtype(data[c]):
        n+=1

print(f"numerical variables: {n}, and none-numerical variables: {len(data.columns) - n}")
```

```python
 data=data[data['Volume']>100000]
```

```python
#Remove all columns that have a constant/single value
data.head(20)
```

```python
constant_columns=[]

for c in data.columns:
    n = len(data[c].unique())
    if n ==1:
        constant_columns.append(c)
        
constant_columns
```

```python
data.drop(columns=constant_columns, inplace=True,axis=1)
```

```python
len(data.columns)
```

```python
data.head(10)
```

```python
 data.info()
```

```python
dateObject = pd.DatetimeIndex(data['Date'])
dateObject
```

```python
data['Day'] = dateObject.day
data['Month'] = dateObject.month
data['Year'] = dateObject.year
```

```python
 data[['Date', 'Day', 'Month', 'Year']]
```

```python
data.columns
```

```python
 # Use visualization for data analysis
# show the trend of stocks volume over years

yearGroup = data[['Year', 'Volume']].groupby('Year')
yearGroup 
```

```python
yearGroup.sum()
```

```python
 yearGroup.count()
```

```python
yearGroup.sum().plot(y='Volume', title='Stocks volume over years')

plt.ylabel('Stocks Volume')
```

```python
monthGroup = data[['Month', 'Adj Close']].groupby('Month')
monthGroup 
```

```python
monthGroup.mean().plot(y='Adj Close', title="Adj Close Over Months")
plt.ylabel('Adj CLose Prices')
```

```python
dayGroup = data[['Day', 'Volume']].groupby('Day')
dayGroup
```

```python
dayGroup.sum()
```

```python
dayGroup.mean().plot(y='Volume', title='Stocks Volume Over Days')
```
