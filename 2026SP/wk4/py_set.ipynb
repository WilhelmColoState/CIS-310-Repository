### Python - Sets, {}, set()

- A set is a collection that is unordered and unindexed, and has no duplicate items
- Unordered - Items do not have a defined order, items cannot be referred to by index or key
- Unchangable - Cannot update the items after the set is created
- no duplicates - no two items with the same value
