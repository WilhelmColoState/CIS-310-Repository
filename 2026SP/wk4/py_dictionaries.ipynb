### Python - Dictionaries, {}, dict()

- Dictionary is a collection that is unordred and changeable
- Dictionaires store data values in key:value pairs
- No duplicate keys
- changable - can change/update, add or remove items after the dictionary is created

```python
#nested dictionary
myFamily = {
    
    'child1': {'name': 'Jen', 'DoB': '01/01/2000'},
    'child2': {'name': 'Bob', 'DoB': '06/01/2004'},
    'spouse': {'name': 'Jon', 'DoB': '12/20/1978', 'MD': '08/08/1998'},
}
```

```python
len(myFamily)
```

```python
myFamily['child1']['name']
```

```python
myFamily.get('child2').get('DoB')
```

```python
myFamily.keys()
```

```python
myFamily['spouse'].keys()
```
