```python
for i in range(colors):
    print(i)
```

```python
for i in range(colors):
    print(colors[i])
```

```python
for i in range(len(colors):
    print(i)
```

```python
for i, c in enumerate(colors):
    print(i,c)
```

```python
for i, c in enumerate(colors, start=1):
    print(i,c)
```

```python
#loop over multiple colloections
#without nested loop
colors = ['red', 'green', 'blue', 'purple', 'yellow']
ratio = [0.2, 0.3, 0.4, 0.2, 0.5]
for j in range(len(colors)):
    print(f"{ratio[j] * 100}% {colors[j]}")
```

```python
# break and continue
for c in colors:
    print(c)
```

```python
for c in colors:
    if c == 'blue':
        break
    print(c)
```

```python
for c in colors:
    if c == 'blue':
        continue
    print(c)
```
